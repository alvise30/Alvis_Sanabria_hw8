#Todo el proceso de Carga de imagen

```python
img = imread('guv_spim0107.tif')
plt.title(u'Imagen Original')
imshow(img,cmap='gray')
```

```python
#Modelo y funciones
def loglikelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum((y_obs-y_model)**2)
    return -chi_squared

def my_model(theta,r0,r1,phi,m):
    return r0 + (r1 * sin((m*theta)+phi))
```

```python
#Crea las caminatas vacías
r0_walk= empty((0))
r1_walk=empty((0))
m_walk=empty((0))
phi_walk=empty((0))

x_walk = empty((0))
y_walk = empty((0))
l_walk = empty((0))
```

```python
r0_walk = append(r0_walk, 80)
r1_walk = append(r1_walk, 1)
m_walk = append(m_walk, 10)
phi_walk = append(phi_walk, 0.1)

x_walk = append(x_walk, 80)
y_walk = append(y_walk, 80)

s_img = shape(img)
theta = linspace(0,2*pi,5000)
comp = zeros((s_img[0],s_img[1]))
for p in range (0,size(theta)):
    j=int(x_walk[0] + sin(p)*my_model(p,r0_walk[0],r1_walk[0],m_walk[0],phi_walk[0]))
    i=int(y_walk[0] + cos(p)*my_model(p,r0_walk[0],r1_walk[0],m_walk[0],phi_walk[0]))
    if j<(shape(img)[0]) and i <(shape(img)[1]):
        comp[j,i]=1

l_walk = append(l_walk,loglikelihood(img,comp))
```

```python
n_iterations = 5000
for i in range(n_iterations):
    if(i%100==0):
        print i,
    r0_prime = random.normal(r0_walk[i], 0.11)
    r1_prime = random.normal(r1_walk[i], 0.1)
    m_prime = random.normal(m_walk[i], 0.1)
    phi_prime = random.normal(phi_walk[i], 0.1)
    
    x_prime = random.normal(x_walk[i], 1)
    y_prime = random.normal(y_walk[i], 1)
    
    theta = linspace(0,2*pi,5000)
    comp1 = zeros((s_img[0],s_img[1]))
    pox = int_(cos(theta)*my_model(theta,r0_prime,r1_prime,phi_prime,m_prime))
    poy = int_(sin(theta)*my_model(theta,r0_prime,r1_prime,phi_prime,m_prime))
    comp1[pox,poy] = 1
    
    l_prime = loglikelihood(img, comp1)
    l_init = l_walk[i]
    
    alpha = exp(l_prime-l_init)
    if(alpha>=1.0):
        r0_walk = append(r0_walk,r0_prime)
        r1_walk = append(r1_walk, r0_prime)
        m_walk = append(m_walk,m_prime)
        phi_walk = append(phi_walk,phi_prime)
        x_walk = append(x_walk, x_prime)        
        y_walk = append(y_walk,y_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta<=alpha):
            r0_walk = append(r0_walk,r0_prime)
            r1_walk = append(r1_walk, r0_prime)
            m_walk = append(m_walk,m_prime)
            phi_walk = append(phi_walk,phi_prime)
            x_walk = append(x_walk, x_prime)        
            y_walk = append(y_walk,y_prime)
            l_walk = append(l_walk, l_prime)
        else:
            r0_walk = append(r0_walk,r0_walk[i])
            r1_walk = append(r1_walk, r0_walk[i])
            m_walk = append(m_walk,m_walk[i])
            phi_walk = append(phi_walk,phi_walk[i])
            x_walk = append(x_walk, x_walk[i])        
            y_walk = append(y_walk,y_walk[i])
            l_walk = append(l_walk, l_walk[i])
```

```python
max_likelihood_id = argmax(l_walk)
best_r0 = r0_walk[max_likelihood_id]
best_r1 = r1_walk[max_likelihood_id]
best_phi = phi_walk[max_likelihood_id]
best_m = m_walk[max_likelihood_id]
best_x = x_walk[max_likelihood_id]
best_y = y_walk[max_likelihood_id]
print l_walk[max_likelihood_id]
print best_r0,best_r1,best_m,best_phi,best_x,best_y
```
