```python
%pylab inline
filee=genfromtxt("antofagasta.dat",delimiter="")
longitud =filee[:,1]
latitud= filee[:,0]
valores =zeros(31)
a = zeros((31,2) )
vectores =zeros( (31,4))
desY =filee[:,4]
desX =filee[:,2]
Y =(longitud/180.0)*6378000* pi
X= (latitud/180.0)*6371000* pi
#como parentesis se implemento X=radio*theta(rads)
for i in range(31):
 matrix=zeros((2,2))
 matrix[0,0]=desX[i]/X[i] 
 matrix[1,1]=desY[i]/Y[i] 
 matrix[0,1] =(1.0/2.0)*( desX[i]/Y[i] + desY[i]/X[i] )
 matrix[1,0]= matrix[0,1]
 valoresP, vectoresP= eig(matrix)
    
 if valoresP[0]> valoresP[1]:
   valores[i] = valoresP[0]
   vectores[i,0] =latitud[i]
   vectores[i,1] =longitud[i]
   vectores[i,2] =100.0*vectoresP[:,0][0] +latitud[i]
   vectores[i,3] =100.0*vectoresP[:,0][1]+longitud[i] 
   a[i,0]=vectoresP[:,0][0]
   a[i,1]= vectoresP[:,0][1]        
 else:
   valores[i]= valoresP[1]
   vectores[i,0]= latitud[i]
   vectores[i,1]= longitud[i]
   vectores[i,2]= (100.0*vectoresP[:,1][0])+latitud[i] 
   vectores[i,3]= (100.0*vectoresP[:,1][1])+longitud[i] 
   a[i,0]= vectoresP[:,1][0]
   a[i,1]= vectoresP[:,1][1]
        
print valores
```
